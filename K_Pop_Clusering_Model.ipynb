## Exploratory Data Analysis


membuat function untuk countplot fitur pada dataset

```python
def count_plot(data, feature):
  fig, ax1 = plt.subplots(figsize=(20,10))
  graph = sns.countplot(ax=ax1,x=feature, data=data)
  i=0
  total = len(data)
  for p in graph.patches:
      height = p.get_height()
      graph.text(p.get_x()+p.get_width()/2., height + 0.2,
          "%.2f" % (data[feature].value_counts()[i]/total * 100) + '%',
          ha="center")
      i += 1
```

# Data Training

```python
from sklearn import metrics
from scipy.spatial.distance import cdist
from sklearn.cluster import KMeans

# gunakan fitur untuk
X = df_kpop[['danceability', 'energy', 'loudness', 'tempo', 'valence', 
             'acousticness']].values
```

# Data Modelling with KMeans

## Finding Optimal K


```python
inertia_values = []
mapping = {}
K = range(1,10)

for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(X)
    inertia_values.append(km.inertia_)
    mapping[k] = km.inertia_

for key, val in mapping.items():
    print(f'{key} : {val}')
```

tampilkan plotnya

```python
plt.plot(K, inertia_values, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method For Optimal k')
plt.show()
```


## Model Implementation
Menggunakan Cluster K = 4

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
```

## KMeans Model Evaluation

### Silhouette Score
Silhouete score of each features scaled

```python
kmeans_silh = metrics.silhouette_score(X, kmeans.labels_)
print(f'Silhouete Score: {kmeans_silh:.2f}')
```

### Davies-Bouldin Index
Davies-Bouldin Index of each features scaled

```python
kmeans_dbi = metrics.davies_bouldin_score(X, kmeans.labels_)
print(f'DBI Score: {kmeans_dbi:.2f}')
```

## View KMeans Label

```python
kmeans_label = kmeans.predict(X)
print(f'KMeans Labels : \n{kmeans_label}')
```

View Unique Value from label

```python
# return count merupakan jumlah fitur dari tiap label
print(f'Unique Values from Labels : {np.unique(kmeans_label, return_counts=True)}')
```

## Find Liveness Average from each Cluster
hitung rata-rata liveness dari tiap cluster dan cari yang tertinggi

```python
df_kpop1 = df_kpop[df_kpop['cluster'] == 0]
df_kpop2 = df_kpop[df_kpop['cluster'] == 1]
df_kpop3 = df_kpop[df_kpop['cluster'] == 2]
df_kpop4 = df_kpop[df_kpop['cluster'] == 3]

df_liveness = pd.DataFrame({'Cluster': ['Cluster 1', 
                                       'Cluster 2', 
                                         'Cluster 3', 
                                         'Cluster 4'],
                            'Average Liveness': [df_kpop1['liveness'].mean(), 
                                                df_kpop2['liveness'].mean(), 
                                                df_kpop3['liveness'].mean(), 
                                                df_kpop4['liveness'].mean()]})

df_liveness['Cluster Rank'] = df_liveness['Average Liveness'].rank(method='dense', ascending=False)
df_liveness['Cluster Rank'] = df_liveness['Cluster Rank'].map({1.0 : 'Many', 2.0 : 'Medium', 3.0 : 'Enough', 4.0 : 'Few'})
df_liveness['Percentage'] = (df_liveness['Average Liveness'] / df_liveness['Average Liveness'].sum()) * 100
# pembulatan bilangan desimal jadi 2 angka dibelakang koma
df_liveness[['Average Liveness', 'Percentage']] = np.round(df_liveness[['Average Liveness', 'Percentage']], 3
                                                           )
                            
df_liveness.sort_values(by='Percentage', ascending=False)
```

```python
plt.figure(figsize=(6,6))
plt.pie(df_liveness['Percentage'], labels = df_liveness['Cluster'], autopct='%1.1f%%')
plt.title('Percentage of Cluster Based on Number of Streams',loc='center', fontsize=15, color='blue')
plt.show()
```
